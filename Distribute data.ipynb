```python
import os
import cv2
import glob
```

```python
data_dir = './data/data_image_folder'
save_dir = './data/white_data'

if not os.path.exists(save_dir):
    os.mkdir(save_dir)
```

```python
data_files = os.listdir(data_dir)
n = 0
for files in data_files:
    date = files[6:10]
    n = files[-4:]
    #print('data=', date)
    #print('n=', n)
    
    img_path = os.path.join(data_dir, files, '0.jpg')
    img = cv2.imread(img_path)
    #print('img_path=', img_path)
    #print(img.shape)
    
    img_name = date+'_'+n+'.jpg'
    img_path = os.path.join(save_dir, img_name)
    print('save_path=', img_path)
    cv2.imwrite(img_path, img)
    
```
